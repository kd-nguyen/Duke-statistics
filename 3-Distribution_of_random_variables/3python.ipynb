```python
from matplotlib import pyplot as plt
from math import factorial
from decimal import Decimal

def prob_binomial(n, k, p):
    return Decimal(factorial(n)) * Decimal(p**k) * Decimal((1 - p)**(n - k)) / (Decimal(factorial(k)) * Decimal(factorial(n - k)))
            
n = 100
p = 0.3
k_values = range(n + 1)
probs_k = [prob_binomial(n, k, p) for k in k_values]
plt.plot(k_values, probs_k)
plt.show()
```

```python
import scipy.stats
def prob_normal(n, mean, stdev):
    z = (n - mean) / stdev
    return scipy.stats.norm.pdf(z)
```
